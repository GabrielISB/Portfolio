```python
import numpy as np
import pandas as pd
import collections 
import itertools 
from itertools import combinations
```

```python
dataset = [
            ['bread', 'milk'],
            ['bread', 'diaper', 'beer', 'egg'],
            ['milk', 'diaper', 'beer', 'cola'],
            ['bread', 'milk', 'diaper', 'beer'],
            ['bread', 'milk', 'diaper', 'cola'],
        ]
```

```python
dataset2 = [
    ["leche","jamón","huevos","Arroz","queso","Yogurt","Coke"],
    ["Arroz","huevos","jamón","queso","Coke","salsa","sandia","Cebolla"],
    ["queso","huevos","Yogurt","salsa","Arroz","Coca cola","Naranja"],
    ["jamón","queso","Cebolla","tomates","sandia","huevos"],
    ["queso","jamón","salsa","Arroz","Limón","Cebolla","Coca cola"],
    ["tomates","Yogurt","Arroz","salsa","Coca cola","tomates","queso"],
    ["Yogurt","huevos","jamón","tomates","Naranja","Arroz","sandia"],
    ["Coca cola","jamón","Yogurt","salsa","sandia"],
]
```

```python
candidates = []
candidates = {}

def support(data):

    for item in data:
        for product in item:
            if product not in candidates.keys():
                candidates[product] = 1
            else:
                candidates.update({product :candidates[product] +1})
    
    return candidates
```

```python
lista = []
lista2 = []
lista3 = []
lista4 = []

def apriori(dataset, minsup):
    sup_1 = support(dataset)
    print("this is the min support", minsup)
    print("1.-")
    print("THESE ARE THE CANDIDATE ITEMS (and their support) IN THE TRANSACTION LIST ")
    print("IN ORDER TO SEE IF THEY CAN BE CONSIDERED AS FREQUENT\n")
    print(sup_1, "\n")
    #frequent itemset 
    for product, value in sup_1.items():
        
        if value >= minsup:
            
            lista.append(product)
    
    print("THESE ARE THE ITEMS THAT COVER MINIMUM SUPPORT IN TRANSACTION LIST: \n")
    print(lista, "\n")
#Aqui es donde saco los pares     
    comb = list(combinations(lista,2))
    print("2.- ")
    print(" \n THESE ARE THE CANDIDATES OF PAIRS \n")
    
    print(comb, "\n")
#Aqui combierto los pares a un diccionario          
    pair_dict = {}  
    for pair in comb:
        string = str(pair)
        pair_dict[string] = 0
#Aqui es donde saco el sporte de los pares         
    for pair in comb:
        for item in dataset:
            if pair[0] in item and pair[1] in item:
                string = str(pair)
                pair_dict[string] = pair_dict[string]+1
    print("THESE ARE THE SUPPORTS OF THE PAIR CANDIDATES \n")
    print(pair_dict, "\n")   
# aqui saco frequent itemset
    for product, value in pair_dict.items():
        
        if value >= minsup:
            lista2.append(product)
            
    print("THESE ARE THE PAIR CANDIDATES THAT COVER MINIMUM SUPPORT \n")        
    print(lista2, "\n")

#aqui es la combinación de tercias
    comb3 = list(combinations(lista,3))
#Tercias a diccionario
    print("3.- ")
    print("\n THESE ARE THE CANDIDATES OF THIRDS \n")
    print(comb3, "\n")
#Combierto tercias a diccionario
    three_dict = {}  
    for three in comb3:
        string = str(three)
        three_dict[string] = 0
#aquí saco el soporte de las tercias 
    for third in comb3:
        for item in dataset:
            if third[0] in item and third[1] in item and third[2] in item:
                string = str(third)
                three_dict[string] = three_dict[string]+1
    
    print("THIS IS THE SUPPORT OF THE THIRDS CANDIDATES \n")
    print(three_dict, "\n")
#aqui saco el frequent itemset   
    for product, value in three_dict.items():
        
        if value >= minsup:
            
            lista3.append(product)
            
    print("THESE ARE THE THIRD CANDIDATES THAT COVER MINIMUM SUPPORT \n")
    print(lista3, "\n")
#Hago la combinación de 4   
    comb4 = list(combinations(lista,4))
    print("4.-")
    print("\n THESE ARE THE CANDIDATES OF FOUR\n")
    print(comb4, "\n")

#aqui combierto las combinaciones de 4 a diccionario

    four_dict = {}
    for four in comb4:
        string = str(four)
        four_dict[string] = 0
#aqui sacare el soporte de las combinaciones de 4

    for four in comb4:
        for item in dataset:
            if four[0] in item and four[1] in item and four[2] in item and four[3] in item:
                string = str(four)
                four_dict[string] = four_dict[string]+1
                
    print("THIS IS THE SUPPORT OF THE CANDIDATES OF FOUR \n")
    print(four_dict, "\n")    
                
    for product, value in four_dict.items():
        if value >= minsup:
            lista4.append(product)
            
    print("THESE ARE THE CANDIDATES OF FOUR THAT COVER MINIMUM SUPPORT \n")
    print(lista4, "\n")
            
        

    
            
            
            
   # return candidates
    
```
