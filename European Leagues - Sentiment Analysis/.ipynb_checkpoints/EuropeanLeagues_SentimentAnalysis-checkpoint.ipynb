## _Twitter Sentiment Analysis about Football European Leauges_

On this sentiment analysis we will analyse the effect European Leagues provoke on people, therefore, there will be a big approach on trying to get insights of what people think about each of these leagues. In addition, we have excluded Ligue UberEats (French League) due to the small number of tweets we could fetch. Despite the fact of the French ligue, we have included Champions League which includes teams from each country. 

The tools used for fetching were: 

#### Twitter Archiver & Tweepy

The main objective of this practice is to do all the stages of data pre processing that are Cleaning, Integration, Transformation, Reduction, including an small analysis of positivity, negativity and neutral comments on each league. For this practice we used the following hashtags: 

#### #SerieA - #LaLiga - #Bundesliga - #PremierLeague

# 1. Data Extraction and Data Cleaning 

We imported both datasets and filtered them in order to have the same columns on both panda dataframes so we could perform the integration. We also excluded columns like Location and Dates as they were not important for this specific matter as we will only analyse the sentiment people have on each league hashtag. In addition we checked out the information of each data frame. For the CSV called "champions" we had to do a change in the names of attributes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
tweets = pd.read_csv("EuropeanLeagues_Tweets.csv", skiprows = 1)
tweets.head()
```

```python
champions.isna().sum()
```

```python
champions=champions[['Tweet ID', 'Tweet Text', 'Full Name', 'Screen Name', 'Favorites','Retweets']]
```

```python
champions.columns
```

```python
tweets = tweets[['Tweet ID', 'Tweet Text', 'Full Name', 'Screen Name', 'Favorites','Retweets']]
```

# 2. Data Integration

Here we added up both data sets in order to obtain a single panda dataframe with all the tweets 

```python
tweets=pd.concat([tweets,champions])
```

# 3. Data Transformation

On the stage of transformation we decided to transform all the tweets we had so we don't find any further error when applying the sentiment analysis. You can see that function below on the called process_tweet(). After that tweets transformation, we have got the tweet without any further symbol mistake. 

```python
def process_tweet():
    #return ",".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])", " ",tweet.lower().split()))
    #train_data['processed_tweets'] = train_data['tweet'].str.replace('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])', '')
    tweets['processed_tweets'] = tweets['Tweet Text'].str.replace('[^A-Za-z0-9 ]', '')
    tweets['processed_tweets'] = tweets['processed_tweets'].str.replace('@[a-zA-Z0-9]+', '')
```

```python
process_tweet()
```

# 4. Data Reduction

On this data reduction stage we reduced the data when we quit the Nan and empty values besides erradicating duplicated tweets as we considered them noisy data. We only wanted to drop all the rows and columns with empty values so we do not have to worry about those kind of data affecting our sentiment analysis. After that we divided the dataframe into a differnet data set for each European League, so once we had data reduced in different data frames we were able to start perfoming the sentiment analysis. 

```python
tweets = tweets.dropna(how='all')
tweets = tweets.dropna(axis = 'columns', how='all')
tweets = tweets.drop_duplicates(subset=['Tweet Text'], keep=False)
```

```python
tweets.info()
```

```python
tweets.head(6)
```

```python
strings=['#SerieA','#SERIEA','#seriea','#serieA','#Seriea']
serie_a=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
serie_a=tweets[serie_a]


strings=['#BundesLiga','#bundesliga','#Bundesliga','#BUNDESLIGA','bundesLiga']
bundesliga=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
bundesliga=tweets[bundesliga]

strings=['#LaLiga','#laliga','#LALIGA','#laLiga','#laLiga']
laliga=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
laliga=tweets[laliga]

strings=['#UberEatsLeague', '#LeagueUberEats','#leagueubereats','#Leagueubereats']
uber=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
uber=tweets[uber]

strings=['#PremierLeague','#premierleague','#Premierleague','#PRMIERLEAGUE','#premierLeague']
premier=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
premier=tweets[premier]

strings=['#ChampionsLeague','#championsleague']
champions=tweets['Tweet Text'].str.contains('|'.join(strings),regex=True)
champions=tweets[champions]

```

```python
champions.info()
```

# __SERIE A__

```python
positive = 0
negative = 0
neutral = 0
polarity = 0
tweet_list = []
neutral_list = []
negative_list = []
positive_list = []

for tweet in serie_a['Tweet Text']:
 
    print(tweet)
    tweet_list.append(tweet)
    analysis = TextBlob(tweet)
    score = SentimentIntensityAnalyzer().polarity_scores(tweet)
    neg = score['neg']
    neu = score['neu']
    pos = score['pos']
    comp = score['compound']
    polarity += analysis.sentiment.polarity

    if neg > pos:
        negative_list.append(tweet)
        negative += 1
    elif pos > neg:
        positive_list.append(tweet)
        positive += 1

    elif pos == neg:
        neutral_list.append(tweet)
        neutral += 1
```

```python
noOfTweet=9333

def percentage(part,whole):
 return 100 * float(part)/float(whole)
 
positive = percentage(positive, noOfTweet)
negative = percentage(negative, noOfTweet)
neutral = percentage(neutral, noOfTweet)
polarity = percentage(polarity, noOfTweet)
positive = format(positive, '.1f')
negative = format(negative, '.1f')
neutral = format(neutral, '.1f')
```

```python
tweet_list = pd.DataFrame(tweet_list)
neutral_list = pd.DataFrame(neutral_list)
negative_list = pd.DataFrame(negative_list)
positive_list = pd.DataFrame(positive_list)
print('total number: ',len(tweet_list))
print('positive number:' ,len(positive_list))
print('negative number:' , len(negative_list))
print('neutral number: ',len(neutral_list))


```

```python
labels = ['positive ['+str(positive)+'%]' , 'Neutral ['+str(neutral)+'%]','Negative ['+str(negative)+'%]']
sizes = [positive, neutral, negative]
colors = ['yellowgreen', 'blue','red']
patches, texts = plt.pie(sizes,colors=colors, startangle=90)
plt.style.use('default')
plt.legend(labels)
plt.title('Sentiment Analysis Result for Serie A' )
plt.axis('equal')
plt.show()
```

```python
fav = serie_a['Favorites']
fav_num = fav.value_counts()

rt = serie_a['Retweets']
rt_num = rt.value_counts()
```

# Conclusions 

_Once we have finished the sentiment analysis we could see that the most tweets on each data set tend to have a "neutral" feeling on what they posted about those leagues. We can clearly see that when we analyzed the tweet with most Retweets and Favorites as some of them had nothing to do directly with the leagues but what happend around it._ 
_On the other hand, we were also able to identify that the tendency is of positive comments on the tweets about each league, however, it does not necessarily mean that negative comments did not appear, as sometimes they almost reached out the same percentage of positive tweets, for example in Bundesliga. On the opposite, if we observe the sentiment percentages on #ChampionsLeague, there is a notice difference between the postiveness and negativeness, something that absolutely make sense as Champions League is the most watched European League worldwide. Another strong example of a notable difference from positiveness and negativeness appears in Premier League Analysis. In fact, Premier League has been called the best league in the world. (Excluding Champions League as it covers five contries), the same happens with Serie A. Finally, La Liga analysis showed up that Messi departure affected on how people see spanish league, making it clear with just a small difference between the feelings in the tweets._
