```python
#Lo volví a abrir

conn = psycopg2.connect(host="localhost", database="BigData_1", user="postgres", password="dancingdead27")
```

```python
#También lo volví a abrir

cur = conn.cursor()
```

```python
#Este codigo es para checar la tabla de POSTGRESQL

cur.execute('SELECT * from "t_t_alumnos"')
result = cur.fetchall() ;
print(result)
```

```python
#Con esta función hice el cambio (UPDATE) de Números a letras NC, Basic, Good, Hero
# result[i][2] hace referencia a la tercer columna de la tupla que da el codigo de arriba

for i in range(20):
    sql2 = 'UPDATE t_t_alumnos SET calificacion = %s WHERE id = %s '
    
    if int(result[i][2]) < 70:
        datos=("NC", i)
        
    elif int(result[i][2]) >= 70 and int(result[i][2]) < 80:
        datos=("Basic", i)
        
    elif int(result[i][2]) >= 80 and int(result[i][2]) < 90:
        datos=("Good", i)
        
    else:
        datos=("Hero", i)
        
    cur.execute(sql2, datos)
    conn.commit()
```
